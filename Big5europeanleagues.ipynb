```python
import pandas as pd
import re

# Load your file
df = pd.read_csv('/content/big5 (1).csv')

# 1. Remove repeated header rows
df = df[df['Rk'] != 'Rk'].copy()

# 2. Drop unnecessary unnamed column if it exists
if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True)

# 3. Fix player names (remove digits + decode weird characters)
df['Player'] = df['Player'].astype(str).str.replace(r'\d+', '', regex=True)
df['Player'] = df['Player'].str.encode('latin1').str.decode('utf-8', errors='ignore')
df['Player'] = df['Player'].apply(lambda x: re.sub(r'[^a-zA-Z\s\-\.]', '', x).strip())  # preserve names like "João Félix" if needed

# 4. Clean nations (keep only 3-letter uppercase code)
df['Nation'] = df['Nation'].str.extract(r'([A-Z]{3})')

# 5. Clean squad names (remove numbers)
df['Squad'] = df['Squad'].str.replace(r'^\d+\s*', '', regex=True)

# 6. Clean league names (standardize full names)
df['Comp'] = df['Comp'].replace({
    'Serie': 'Serie A',
    'La': 'La Liga',
    'Premier': 'Premier League',
    'Ligue': 'Ligue 1',
    'Bundes': 'Bundesliga'
})

# 7. Drop 'Matches' or similar AM columns (in case it's mislabeled)
df = df.loc[:, ~df.columns.str.contains('Matches|AM', case=False)]

# 8. Reset index
df.reset_index(drop=True, inplace=True)

# Preview
df.head()
```

```python
import pandas as pd
import re
from unidecode import unidecode

# Load the CSV (adjust the path if needed)
df = pd.read_csv('/content/big5 (1).csv')

# 1. Remove repeated header rows
df = df[df['Rk'] != 'Rk'].copy()

# 2. Drop the 'Rk' column
df.drop(columns=['Rk'], inplace=True)

# 3. Clean player names
df['Player'] = df['Player'].astype(str).str.replace(r'\d+', '', regex=True)
df['Player'] = df['Player'].apply(lambda x: unidecode(x).strip())  # e.g., Željko → Zeljko
df['Player'] = df['Player'].apply(lambda x: re.sub(r'[^a-zA-Z\s\-\.]', '', x).strip())

# 4. Clean nation values (keep only uppercase part)
df['Nation'] = df['Nation'].astype(str).apply(lambda x: x.split()[-1])

# 5. Clean club names
df['Squad'] = df['Squad'].astype(str).apply(lambda x: re.sub(r'\d+', '', x).strip())

# 6. Normalize competition names
df['Comp'] = df['Comp'].replace({
    'Serie': 'Serie A',
    'La': 'La Liga',
    'Premier': 'Premier League'
})

# 7. Clean 'MP' column (remove "AM" if present)
if 'MP' in df.columns:
    df['MP'] = df['MP'].astype(str).str.replace('AM', '', regex=False).str.strip()

# 8. Display the cleaned data
df.head()
```
